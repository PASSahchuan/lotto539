```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import matplotlib.pyplot as plt
```

```python
#紀錄近半年中獎數字
number = []
#紀錄近半年中獎數字總數
count = 0
```

```python
#取得目前年份、月份
year = int(datetime.now().strftime('%Y'))-1911
month = int(datetime.now().strftime('%m'))
print(year)
print(month)
```

```python
#使用模擬瀏覽器開啟此網站
chrome = webdriver.Chrome('/Users/chuanlin/chromedriver')
chrome.get("https://www.taiwanlottery.com.tw/lotto/dailycash/history.aspx")
time.sleep(5)

#勾選選擇使用年月來篩選需要的資料
selectBtn = chrome.find_element_by_id('D539Control_history1_radYM')
selectBtn.click()
time.sleep(5)
```

```python
#click button into month
for i in range(6):
    nowMonth = month - i
    if nowMonth <= 0:
        nowMonth = nowMonth + 12
        nowYear = year-1
    else:
        nowYear = year
    selectBtn = chrome.find_element_by_id('D539Control_history1_radYM')
    selectBtn.click()
    time.sleep(5)

    #選擇年份以及月份
    selectYear = Select(chrome.find_element_by_id('D539Control_history1_dropYear'))
    selectYear.select_by_value(str(nowYear))
    time.sleep(5)
    selectMonth = Select(chrome.find_element_by_id('D539Control_history1_dropMonth'))
    selectMonth.select_by_value(str(nowMonth))
    time.sleep(5)

    #按下搜尋鍵
    searchBtn = chrome.find_element_by_id('D539Control_history1_btnSubmit')
    searchBtn.click()
    time.sleep(5)
    
    print(nowYear)
    print(nowMonth)
    
    soup = BeautifulSoup(chrome.page_source, "html.parser")
    
    for i in range(31):
        dateId = 'D539Control_history1_dlQuery_D539_DDate_' + str(i)
        result = soup.find(id = dateId)
        if result:
            print(result.string)
            for j in range(1, 6):
                numberId = 'D539Control_history1_dlQuery_No'+str(j)+'_'+str(i)
                numberResult = soup.find(id = numberId)
                count = count + 1
                print(numberResult.string)
                number.append(int(numberResult.string))
```

```python
#近半年內所有中獎數字
#print(number[0:count])
```

```python
#將所有中獎數字排序
number.sort()
#print(number[0:count])
```

```python
#中獎數字
prizeNumber = []
#中獎數字近半年內抽中次數
prizeCount = []
d = {}
for i in range(1,40):
    prizeNumber.append(i)
    #print(i)
    prizeCount.append(number.count(i))
    #print(number.count(i))
    d[i] = number.count(i)
print(d)
```

```python
plt.plot(prizeNumber, prizeCount)
plt.show()
```
